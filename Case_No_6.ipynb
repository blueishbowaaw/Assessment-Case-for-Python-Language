```python
case_6 = pd.read_csv("Documents/Data_Science/dataset_superstore_simple.csv", parse_dates = ["order_date"])
case_6
```

```python
case_6["year"] = case_6.order_date.dt.strftime("%Y")
case_6
```

```python
step1 = case_6[["year","sales","profit"]]
step1
```

```python
step2 = step1.set_index("year")
step2
```

```python
step3 = step2.loc[["2014","2015"]].reset_index()
step3
```

```python
import plotly
import plotly.graph_objects as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot

init_notebook_mode(connected = True)
```

```python
plot_result_No_6 = go.Scatter(
    x = step3.sales,
    y = step3.profit,
    mode = "markers",
)
layout = ({"title":{
    "text": "Graph of Sales VS Profit 2014-2015",
    "x":0.5,
    }
}
)
graph_result_No_6 = go.Figure (data = plot_result_No_6, layout = layout)
iplot (graph_result_No_6)
```

```python
import numpy as np
```

```python
step3["marker_color"] = np.where (
    step3.year == "2014",
    "Red",
    "Blue"
)
step3
```

```python
plot2_result_No_6 = go.Scatter(
    x = step3.sales,
    y = step3.profit,
    mode = "markers",
    marker = {"color": step3.marker_color},  
)
layout = ({"title":{
    "text": "Graph of Sales VS Profit 2014-2015",
    "x":0.5,
    },
    "xaxis":{
        "title": "Sales",
    },
    "yaxis":{
        "title":"Profit",
    }
}
)
graph2_result_No_6 = go.Figure (data = plot2_result_No_6, layout = layout)
iplot (graph2_result_No_6)
```
